```python
import pandas as pd
```

1.将上方数据表建立 DataFrame 进行存储

```python
df=pd.DataFrame({'id':['1001','1001','1002','1003','1001','1002','1002','1003','1001','1002','1001','1002'],
                'date':['2019-01-01','2019-01-01','2019-01-01','2019-01-01','2019-01-02','2019-01-02','2019-01-02','2019-01-02',\
                        '2019-01-03','2019-01-03','2019-01-04','2019-01-04'],
                'goods':['A','B','A','B','A','A','B','A','A','B','A','A'],
                'sales':[3000,2000,3000,2000,3000,3000,2000,3000,3000,2000,3000,3000]})
df
```

2.利用 pivot_table 统计该企业每个人每天销售商品 A 和 B 的金额之和

```python
df_pivot=df.pivot_table(index=['id'],columns=['date'],values=['sales'],aggfunc='sum',fill_value=0)
df_pivot
```

3.请将上面的统计数据再增加一列，统计出每个人的销售总金额

```python
df_pivot=df.pivot_table(index=['id'],columns=['date'],values=['sales'],aggfunc='sum',fill_value=0,margins=True,margins_name='sales_sum')
df_pivot
```

4.请将以上数据存入 DataFrame 中

```python
df2=pd.DataFrame([{'sex':'man','age':15,'smoke':True,'height':168},{'sex':'man','age':23,'smoke':False,'height':179},
                  {'sex':'woman','age':25,'smoke':False,'height':181},{'sex':'woman','age':17,'smoke':True,'height':166},
                  {'sex':'man','age':35,'smoke':True,'height':173},{'sex':'woman','age':57,'smoke':False,'height':178},
                  {'sex':'man','age':24,'smoke':False,'height':188},{'sex':'woman','age':31,'smoke':True,'height':190},
                  {'sex':'woman','age':22,'smoke':False,'height':160}])
df2
```

5.请以性别为分组依据，查看男女抽烟平均年龄和身高

```python
df2[df2['smoke']==True].groupby('sex')[['age','height']].mean()
```

6.利用交叉表知识，统计各个性别抽烟的人数

```python
pd.crosstab(df2.sex,df2.smoke)
```

7.利用交叉表知识，统计各个年龄段抽烟人情况

```python
pd.crosstab(pd.cut(df2.age,bins=[14,20,25,30,60]),df2.smoke)
```
