```python
from pyecharts import options as opts
from pyecharts.charts import Liquid,Gauge,WordCloud,Line
#from pyecharts.globals import SymbolType
import tushare as ts
```

```python
df = ts.get_stock_basics()

gpr_avg=round(df['gpr'].mean()/100,4)
npr_avg=round(df['npr'].mean()/100,4)
```

```python
liquid=(
        Liquid()
        .add("",[gpr_avg,npr_avg],shape='circle')
        # 水球外形，有' circle', 'rect', 'roundRect', 'triangle', 'diamond', 'pin', 'arrow' 可选。
        # 默认 'circle'。也可以为自定义的 SVG 路径。
        #is_outline_show设置边框
        .set_global_opts(title_opts=opts.TitleOpts(title='毛利率-净利率水球图'))
        )
liquid.render_notebook()
```

```python
gauge=(
      Gauge()
      .add('毛利率',[('平均数',gpr_avg*100)])
      .set_global_opts(title_opts=opts.TitleOpts(title='毛利率仪表盘图'))
      )
gauge.render_notebook()
```

2.统计 A 股各行业公司数量，并跟进行业分布数据作词云图(阅读参考文章 4、5);

```python
df_ind=df.groupby('industry')['industry'].agg([('公司数量','count')])
words=list(zip(df_ind.index.tolist(),df_ind.values.tolist()))
words
```

```python
wordcloud=(
        WordCloud()
        .add('',words,word_size_range=[20,max(df_ind.values.tolist())],shape='diamod')
        .set_global_opts(title_opts=opts.TitleOpts(title='行业词云图'))
        )
wordcloud.render_notebook()
```

3.利用 tushare 提取茅台数据中的最高最低价，作双折线图，茅台股票代码 600519
 (提示:提取单个股票信息 df = ts.get_hist_data('600519',start=' ',end=' ') )

```python
df3 = ts.get_hist_data('600519').sort_index()
x=df3.index.tolist()

y1=df3['high'].tolist()
y2=df3['low'].tolist()
```

```python
line=(
      Line()
      .add_xaxis(x)
      .add_yaxis('最高价',y1,symbol_size=10)
      .add_yaxis('最低价',y2,symbol='triangle',symbol_size=10)
      .set_series_opts(
              #areastyle_opts=opts.AreaStyleOpts(opacity=0.5), #折线面积
              label_opts=opts.LabelOpts(is_show=False) #标签的显示隐藏
      )
      .set_global_opts(title_opts=opts.TitleOpts(title='茅台每日最高价和最低价'),
                       yaxis_opts=opts.AxisOpts(type_="value",min_=400)
                       )
      )
line.render_notebook()
```
