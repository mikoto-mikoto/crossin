5.请统计出不同户型对应的房源数量、平均面积；并从大到小进行排序，判断出主流户型是什么？平均面积多大？

```python
df_g=df.groupby('huxing')[['title','size']].agg(['count','mean']).reset_index()
df_g.columns=['户型','房源数量','平均面积']
df_g=df_g.sort_values(by='房源数量',ascending=False)
df_g
```

```python
print('主流户型是：',df_g.iloc[0,0],'，平均面积：',df_g.iloc[0,2])
```
